```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

import mbo_utilities as mbo
import lbm_suite2p_python as lsp
```

```python
raw_data = Path(r"D:/demo/raw")
```

```python
arr = mbo.imread(
    inputs=raw_data,  # str | Path | ndarray | Sequence[str | Path]
)
```

```python
print(f"Shape: {arr.shape}")
print(f"Type:  {type(arr).__name__}")

if hasattr(arr, 'metadata'):
    md = arr.metadata
    print(f"Num planes: {md.get('num_planes', 'N/A')}")
    print(f"Frame rate: {md.get('fs', 'N/A')} Hz")
else:
    print("No metadata found for input file")
```

```python
# Below are all defaults
ops = lsp.default_ops()
```

```python
ops['tau'] = 1.3  # gcamp indicator decay time
ops['fs'] = arr.metadata.get('frame_rate', 10.0)
ops['nplanes'] = arr.metadata.get('num_planes', 1)
ops['do_registration'] = True
```

```python
lsp.pipeline(arr, ops=ops, num_timepoints=500)
```
